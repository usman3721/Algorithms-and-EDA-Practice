```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_excel("./Data_Train.xlsx")
df_train.head()
```

```python
df_test=pd.read_excel("./Test_set.xlsx")
df_test
```

```python
final_df=df_train.append(df_test)
final_df.head()
```

```python
final_df.info()
```

```python
# Feature engineering
final_df["Date"]=final_df["Date_of_Journey"].apply(lambda x:x.split("/")[0])
final_df["Month"]=final_df["Date_of_Journey"].apply(lambda x:x.split("/")[1])
final_df["Year"]=final_df["Date_of_Journey"].apply(lambda x:x.split("/")[2])
```

```python
final_df.head()
```

```python
final_df["Date"]=final_df["Date"].astype(int)
final_df["Month"]=final_df["Month"].astype(int)
final_df["Year"]=final_df["Year"].astype(int)
```

```python
final_df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
final_df["Arrival_Time"]=final_df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
```

```python
final_df["Arrival_Hour"]=final_df["Arrival_Time"].apply(lambda x:x.split(":")[0])
final_df["Arrival_Min"]=final_df["Arrival_Time"].apply(lambda x:x.split(":")[1])
```

```python
final_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
final_df["Arrival_Hour"]=final_df["Arrival_Hour"].astype(int)
final_df["Arrival_Min"]=final_df["Arrival_Min"].astype(int)
```

```python
final_df.head(1)
```

```python
final_df["Dep_Time_Hour"]=final_df["Dep_Time"].apply(lambda x:x.split(":")[0])
final_df["Dep_Time_Min"]=final_df["Dep_Time"].apply(lambda x:x.split(":")[1])
```

```python
final_df["Dep_Time_Hour"]=final_df["Arrival_Hour"].astype(int)
final_df["Dep_Time_Min"]=final_df["Arrival_Min"].astype(int)
```

```python
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
final_df.Total_Stops.unique()
```

```python
final_df["Total_Stops"]=final_df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,"nan":2})
```

```python
final_df.drop("Route",axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df["Duration_hour"]=final_df["Duration"].str.split(" ").str[0].str.split("h").str[0]





```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
final_df["Airline"]=encoder.fit_transform(final_df["Airline"])
final_df["Source"]=encoder.fit_transform(final_df["Source"])
final_df["Destination"]=encoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"]=encoder.fit_transform(final_df["Additional_Info"])
```
